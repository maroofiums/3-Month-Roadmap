## Basics of Pandas


```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop("deck",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["age"].fillna(df["age"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["embark_town"] = df["embark_town"].fillna(df["embark_town"].mode()[0])
```
